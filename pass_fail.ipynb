**PROJECT TITLE:** Student Performance Prediction using Logistic Regression.

**OBJECTIVE:**  To build a simple machine learning model that predicts whether a student will pass (1) or fail (0) based on:

Study hours

Attendance (%)

**PROCEDURE/CODE:**

**Step 1: Import Required Libraries**

pandas: For creating and handling data in a table (DataFrame) format.

matplotlib.pyplot: For plotting graphs to visualize data.

sklearn.model_selection.train_test_split: To split data into training and test sets.

sklearn.linear_model.LogisticRegression: For building the logistic regression model.

sklearn.metrics.accuracy_score: To measure the modelâ€™s prediction accuracy.

```python
# Importing all the necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Step 2: Create the Dataset**

A small dataset is created manually using dictionaries.

It has three columns:

Hours_Studied: How many hours the student studied.

Attendance: Their attendance percentage.

Pass_Fail: The result (1 = Pass, 0 = Fail).

```python
# Manually creating a small dataset

data = {
    'Hours_Studied': [5, 2, 4, 1, 6, 3, 2.5, 4.5, 3.5, 1.5],
    'Attendance': [85, 60, 75, 50, 90, 65, 58, 80, 70, 55],
    'Pass_Fail': [1, 0, 1, 0, 1, 1, 0, 1, 1, 0]
}

df = pd.DataFrame(data)
df
```

**Step 3: Visualize the Dataset**

Creating a scatter plot to visualize the relation between hours studied and attendance which helps to understand patterns before training the model.

Green dots represent students who passed and the Red dots show those who failed.

```python
# Scatter plot of Hours Studied vs Attendance

plt.figure(figsize=(8, 5))
for i in range(len(df)):
    color = 'green' if df['Pass_Fail'][i] == 1 else 'red'
    plt.scatter(df['Hours_Studied'][i], df['Attendance'][i], color=color)
plt.xlabel("Hours Studied")
plt.ylabel("Attendance (%)")
plt.title("Student Performance (Green = Pass, Red = Fail)")
plt.grid(True)
plt.show()
```

**Step 4: Prepare Data for Training**

X contains input features (hours + attendance).

y is the output (pass or fail).

Splitting the data into:

Training set (used to train the model)

Testing set (used to test the model's performance)

Usually we use 70% for training and 30% for testing.

```python
X = df[['Hours_Studied', 'Attendance']]   # Features
y = df['Pass_Fail']                       # Target

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

**Step 5: Train the Model**

We create a logistic regression model.

The model is trained on the training set using .fit().

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

**Step 6: Evaluate the Model**

The trained model is used to predict outcomes for the test set.

We compare predictions with actual values using accuracy.

Accuracy tells us how well the model is performing.

```python
# Predict on test data
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

**Step 7: Predict New Values**

We test the model with new student data to see whether they are predicted to pass or fail.

The model uses .predict() to output the result.

```python
# Test with new inputs
new_data = pd.DataFrame({
    'Hours_Studied': [1.5, 5.5],
    'Attendance': [50, 88]
})

predictions = model.predict(new_data)

# Display the predictions
for i in range(len(new_data)):
    status = "Pass" if predictions[i] == 1 else "Fail"
    print(f"Student {i+1}: Hours = {new_data['Hours_Studied'][i]}, Attendance = {new_data['Attendance'][i]} âž¡ Prediction: {status}")
```

Visualize the relationship between study habits and performance,

Train a logistic regression classifier,

And make predictions on new data.

The model demonstrated that students with higher study hours and better attendance are more likely to pass which aligns with real-world expectations.
